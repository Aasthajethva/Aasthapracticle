```python
'''Write a Pandas program to implement following operation
 to convert a NumPy array to a Pandas series
 to convert the first column of a DataFrame as a Series
 to create the mean and standard deviation of the data of a given 
Series
 to sort a given Series'''
```

```python
import numpy as np
import pandas as pd
arr=np.array([1,2,3,4,5])
print("NumPy array:")
print(arr)
print("Converted Pandas series:")
print(pd.Series(arr))
```

```python
import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Hadoop","Python","PySpark"],
    'Fee' :[20000,25000,26000,22000,24000],
    'Duration':['30days','40days','35days','40days','60days'],
    'Discount':[1000,2300,1200,2500,2000]
              }
df = pd.DataFrame(technologies)
print(df)

# Get first column as a series
df2 = df.iloc[:, 0]
print(df2)
```

```python
s = pd.Series(data = [1,2,3,4,5,6,7,8,9,5,3])
print("Original Data Series:")
print(s)
print("Mean of the said Data Series:")
print(s.mean())
print("Standard deviation of the said Data Series:")
print(s.std())
```

```python
import pandas as pd
s = pd.Series(['100', '200', 'python', '300.12', '400'])
print("Original Data Series:")
print(s)
new_s = pd.Series(s).sort_values()
print("Sorted Series:")
print(new_s)
```
